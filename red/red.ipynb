# EU renewable energy rules – feedback review

EU’s Green Deal might lead to changes in the current RED. EC consultation's 
 [feedback](https://ec.europa.eu/info/law/better-regulation/have-your-say/initiatives/12553-EU-renewable-energy-rules-review) analysis (few descriptive statistics).

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('red.csv', low_memory=False)
```

```python
df = df.fillna('n/a')
```

```python
df.columns
```

```python
pd.set_option('display.max_colwidth', 50)
```

```python
df.head()
```

## Feedback by subject type

```python
df['userType'].value_counts()
```

```python
ax = df['userType'].value_counts().plot(
    kind='bar',
    title='Feedback by subject type',
    ylabel='Count',
    )
fig = ax.get_figure()
fig.tight_layout()
fig.savefig('user.svg')
```

## Feedback by country


```python
df['country'].value_counts()
```

```python
ax = df['country'].value_counts().plot(
    kind='bar',
    title='Feedback by country',
    ylabel='Count',
    )
fig = ax.get_figure()
fig.tight_layout()
fig.savefig('country.svg')
```
